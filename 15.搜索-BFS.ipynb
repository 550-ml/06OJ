```python
# 最重要的是怎么建立一个表或者一个图
N = 100010
graph = [0] * N  # 每个位置对应着点的编号，存储着i个点的邻接点
state = [0] * N  # bool数组，存储点的搜索状态

def add(a,b):
    if graph[a]==0:
        graph[a]=[b]
    else:
        graph[a].append(b)
    if graph[b]==0:
        graph[b]=[a]
    else:
        graph[b].append(a)
        
# dfs两种模板
# 第一种是所有可能模板，一般叶子节点才是终点
# 第二种是对图的遍历
def dfs(u):
    state[u]=1
    for i in graph[u]:
        if state[i]==0:
            dfs(i)
```

```python
from collections import deque
import copy

# 输入初始矩阵
line = input().split()
matrix = [[0]*3 for _ in range(3)]
for i in range(3):
    for j in range(3):
        matrix[i][j] = line[i*3+j]
        if matrix[i][j] == 'x':
            start_pos = (i, j)

print("Initial matrix:")
for row in matrix:
    print(row)


def get_pos(matrix_a):
    for i in range(3):
        for j in range(3):
            if matrix_a[i][j] == 'x':
                return (i, j)


def is_goal(new_matrix):
    goal_matrix = [['1', '2', '3'], ['4', '5', '6'], ['7', '8', 'x']]
    return new_matrix == goal_matrix


def matrix_to_tuple(matrix):
    return tuple([tuple(row) for row in matrix])


def bfs():
    step = 0
    dx = [1, -1, 0, 0]
    dy = [0, 0, 1, -1]
    q = deque([copy.deepcopy(matrix)])
    hash_set = set([matrix_to_tuple(matrix)])
    while q:
        size = len(q)
        step += 1
        for _ in range(size):
            current_matrix = q.popleft()
            x, y = get_pos(current_matrix)
            for i in range(4):
                new_x = x + dx[i]
                new_y = y + dy[i]
                if new_x < 0 or new_y < 0 or new_x >= 3 or new_y >= 3:
                    continue
                new_matrix = copy.deepcopy(current_matrix)
                new_matrix[new_x][new_y], new_matrix[x][y] = new_matrix[x][y], new_matrix[new_x][new_y]
                new_tuple = matrix_to_tuple(new_matrix)
                if new_tuple in hash_set:
                    continue
                hash_set.add(new_tuple)
                q.append(new_matrix)
                if is_goal(new_matrix):
                    return step, new_matrix
    return -1, None


# 运行 BFS 算法
step_count, result_matrix = bfs()

if step_count != -1:
    print(f"Solution found in {step_count} steps")
    print("Solution matrix:")
    for row in result_matrix:
        print(row)
else:
    print("No solution found")
```
