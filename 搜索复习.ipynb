```python
from collections import deque
(m,n, t) = map(int, input().split())

matrix = [[0]*n for _ in range(m)]
visited = matrix.deepcopy()
queue = deque()
for i in range(m):
    line = input()
    for j in len(line):
        matrix[i][j] = line[j]
        # 记录起点
        if line[j] == '*':
            start = (i,j)
        if line[j] == '+':
            end = (i,j)

def bfs():
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]
    step = 0
    queue.append(start)
    i,j = start
    visited[i][j] = t

    while queue:
        size = len(queue)
        for _ in range(size):
            x, y = queue.popleft()
            xue = visited[x][y]
            for i in range(4):
                new_x = x+dx[i]
                new_y = y+dy[i]
                if new_x>=0 and new_x<m and new_y>=0 and new_y<n and matrix[new_x][new_y]!='#':
                    
                    if matrix[new_x][new_y] == '.' or matrix[new_x][new_y] =='+':
                        if xue>visited[new_x][new_y]:
                            visited[new_x][new_y] =xue
                            queue.append((new_x,new_y))
                    else:
                        guai = int(matrix[new_x][new_y])
                        
                        if xue-guai>visited[new_x][new_y]:
                            visited[new_x][new_y] =xue-guai
                            queue.append((new_x,new_y))
bfs()
(i,j) = end
print(visited[i][j])
                    
                        
```

```python
# 拓扑排序
(n, m) = map(int, input().split())
# 
graph = {}
ans = []
q = []
p_in = [0]*(n+1)

def add(a,b):
    if a not in graph:
        graph[a] = [b]
    else:
        graph[a].append(b)
    p_in[b]+=1

for _ in range(m):
    (a,b) = map(int,input().split())
    add(a,b)

def top_sort():
    for i in range(1,n+1):
        if p_in[i] == 0:
            q.append(i)
    
    while q:
        x = q.pop(0)
        ans.append(x)
        if x in graph:
            for b in graph[x]:
                p_in[b]-=1
                if p_in[b] == 0:
                    q.append(b)
top_sort
if len(ans) == n:
    print()
```
