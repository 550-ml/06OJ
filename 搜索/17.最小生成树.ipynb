```python
def find(parent, i):
    if parent[i] == i:
        return i
    return find(parent, parent[i])


def union(parent, rank, x, y):
    xroot = find(parent, x)
    yroot = find(parent, y)

    if rank[xroot] < rank[yroot]:
        parent[xroot] = yroot
    elif rank[xroot] > rank[yroot]:
        parent[yroot] = xroot
    else:
        parent[yroot] = xroot
        rank[xroot] += 1


def minimum_cost(n, edges):
    result = 0
    edges.sort(key=lambda item: item[2])  # 按成本排序
    parent = [i for i in range(n+1)]
    rank = [0] * (n+1)

    for edge in edges:
        u, v, w, status = edge
        if status == 0:  # 如果路还没有修建
            uroot = find(parent, u)
            vroot = find(parent, v)

            if uroot != vroot:
                union(parent, rank, uroot, vroot)
                result += w

    return result


while True:
    n = int(input())
    if n == 0:
        break
    edges = []
    for _ in range(n * (n - 1) // 2):
        edges.append(list(map(int, input().split())))
    print(minimum_cost(n, edges))
```
