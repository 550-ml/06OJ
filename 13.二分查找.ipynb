```python
(n, q) = map(int , input().split())
num_list = list(map(int ,input().split()))

# 这实际上式返回第一个等于的位置
def lower_bound(num_list, target):
    left = 0
    right = len(num_list)-1
    while left<= right:
        mid = (left+right)//2
        if num_list[mid]< target:
            left = mid+1
        else:
            right = mid-1
    return left

def get_fanwei(num_list, target):
    start = lower_bound(num_list, target)
    if start == len(num_list):
        return -1,-1
    end = lower_bound(num_list,target+1)-1
    return start, end

for _ in range(q):
    num = int(input())
    start, end = get_fanwei(num_list, num)
    if (end<start):
        start,end = -1,-1
    print(start,end=' ')
    print(end)
```

```python
def find_num(num_list: List[int], target: int) :
    try:
        # 找到目标数字的索引
        index = num_list.index(target)
    except ValueError:
        # 如果目标数字不在列表中，返回 (-1, -1)
        return [-1, -1]

    # 由于列表是有序的，我们可以快速找到目标数字的起始和终止位置
    start = index
    end = index + 1
    while end < len(num_list) and num_list[end] == target:
        end += 1

    # 返回目标数字的起始和终止位置
    return [start, end - 1]
```

```python
# 超时，看来必须写一个前缀和
from typing import List

def create_prefix_list(num_list: List):
    n = len(num_list)
    prefix_list = [0] * n
    prefix_list[0] = num_list[0]
    for i in range(1,n):
        prefix_list[i] = prefix_list[i-1] + num_list[i]
    return prefix_list

def get_num(prefix_list, l, r):
    if l>1:
        ans= prefix_list[r-1]-prefix_list[l-2]
    else:
        ans = prefix_list[r-1]

    return ans
(n, m) = map(int, input().split())

num_list = list(map(int, input().split()))
num_list = create_prefix_list(num_list)

for _ in range(m):
    (l, r) = map(int, input().split())

    print(get_num(num_list, l, r))
```

```python
# 二维子矩阵
def create_prefix_matrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    prefix_matrix = [[0]*(m+1) for _ in range(n+1)]
    # print(prefix_matrix)
    for i in range(1, n+1):
        for j in range(1, m+1):
            prefix_matrix[i][j] = matrix[i-1][j-1]+ prefix_matrix[i-1][j]+ prefix_matrix[i][j-1]- prefix_matrix[i-1][j-1]
    return prefix_matrix
matrix = [[1,2,3],[4,5,6],[7,8,9]]
# print(create_prefix_matrix(matrix))
def get_sub_matrix(prefix_matrix, x1, y1, x2, y2):
    return prefix_matrix[x2][y2]- prefix_matrix[x1-1][y2]- prefix_matrix[x2][y1-1]+ prefix_matrix[x1-1][y1-1]

(n, m, q) = map(int, input().split())
matrix = [list(map(int, input().split())) for i in range(n)]
prefix_matrix = create_prefix_matrix(matrix)
for _ in range(q):
    (x1,y1,x2,y2) = map(int,input().split())
    print(get_sub_matrix(prefix_matrix,x1, y1, x2,y2))
```
