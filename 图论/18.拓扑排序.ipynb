```python
# 重写拓扑排序
(n, m) = map(int, input().split())
# 创建数据结构
graph = {}
p_in = [0] * (n+1)
q = []
ans = []
jishu = n

def add(a, b):
    if a not in graph:
        graph[a] = []
    graph[a].append(b)


for i in range(m):
    (a, b) = map(int, input().split())
    add(a, b)
    p_in[b] += 1


def top_sort():
    global jishu
    for i in range(1, n+1):
        if p_in[i] == 0:
            q.append(i)
    while q:
        x = q.pop(0)
        ans.append(x)
        jishu -= 1
        # 检查
        if x in graph:
            for point in graph[x]:
                p_in[point] -= 1
                if p_in[point] == 0:
                    q.append(point)


top_sort()
if jishu == 0:
    print(' '.join(map(str, ans)))
    
else:
    print(-1)
```
