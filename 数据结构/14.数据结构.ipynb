```python
# 单调栈
n = int(input())
num_list = list(map(int, input().split()))

ans = []
stack = []
for num in num_list:
    if not stack:
        stack.append(num)
        ans.append(-1)
    else:
        # 有元素
        while num <= stack[-1]:
            stack.pop()
            if not stack:
                stack.append(num)
                ans.append(-1)
                break
        else:
            ans.append(stack[-1])
            stack.append(num)
for num in ans:
    print(num, end=' ')
```

```python
from collections import deque
# 单调队列， 滑动窗口
n, k = map(int, input().split())
num_list = list(map(int, input().split()))

# 窗口最小值
min_queue = deque()

for i in range(n):
    # 队列不为空，且当前元素小于队尾元素
    while min_queue and num_list[min_queue[-1]] >= num_list[i]:
        min_queue.pop()
    min_queue.append(i)
    
    if min_queue[0] == i - k:
        min_queue.popleft()
    if i>=k-1:
        print(num_list[min_queue[0]], end=' ')
        
print()
# 
max_queue = deque()
for i in range(n):
    while max_queue and num_list[max_queue[-1]] <= num_list[i]:
        max_queue.pop()
    max_queue.append(i)
    
    if max_queue[0] == i - k:
        max_queue.popleft()
    
    if i >= k - 1:
        print(num_list[max_queue[0]], end=' ')  
```

```python
# 并查集
(n, m ) = map(int, input().split())
p = [0]*(n+1)
for i in range(n+1):
    p[i] = i

def find(p, a):
    if p[a] == a:
        return a
    p[a] = find(p, p[a])
    return p[a]
for _ in range(m):
    input_list = input().split()
    if input_list[0] == 'M':
        a,b =int(input_list[1]), int(input_list[2])
        
        f_a =find(p,a)
        f_b = find(p,b)
        if f_a!=f_b:
            p[a] = f_b
    
    if input_list[0] == 'Q':
        a,b =int(input_list[1]), int(input_list[2])
        
        f_a =find(p,a)
        f_b = find(p,b)
        if f_a==f_b:
            print('Yes')
        else:
            print('No')
```

```python
# 连通块的数量
(n, m) = map(int, input().split())
p = [0]* (n+1)

for i in range(1,n):
    p[i] = i

def find(x):
    if p[x] == x:
        return x
    p[x] = find(p[x])
    return p[x]
for _ in range(m):
    line = input().split()
    if line[0] == 'C':
        a = int(line[1])
        b = int(line[2])
        f_a = find(a)
        f_b = find(b)
        if f_a != f_b:
            p[f_a] = f_b
    if line[0] =='Q1':
        a = int(line[1])
        b = int(line[2])
        f_a = find(a)
        f_b = find(b)
        if f_a == f_b:
            print('Yes')
        else:
            print('No')
    
    if line[0] == 'Q2':
        a = int(line[1])
        f_a = find(a)
        count = p.count(a)
        print(count)
```

```python
# 哈希
# 就是把字符串映射成数字
(n, m) = map(int, input().split())

s = ' '+input()
p = [0]*(n+1)  # 预处理p的多少次幂
h = [0]*(n+1)  # 前l-r的hash
P= 131
# 预处理
p[0] = 1
h[0] = ord(' ')
for i in range(1,n+1):
    p[i] = p[i-1]*P
    h[i] = h[i-1]*P + ord(s[i])
    
def get_num(l, r):
    return h[r] - h[l-1]*p[r-l+1]
for _ in range(m):
    l1,r1,l2,r2 = map(int, input().split())
    if get_num(l1,r1) == get_num(l2,r2):
        print("Yes")
    else:
        print('No')
```
